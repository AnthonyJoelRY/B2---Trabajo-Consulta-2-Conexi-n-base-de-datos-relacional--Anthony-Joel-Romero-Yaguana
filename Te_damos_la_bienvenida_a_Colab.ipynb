```
# ===============================================
# üìä An√°lisis de la variable Global_Sales
# ===============================================
# Este programa:
# 1. Permite seleccionar un archivo CSV con un picker
# 2. Carga los datos en un DataFrame
# 3. Muestra un histograma y un diagrama de caja
# ===============================================

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# --- 1Ô∏è‚É£ Subir archivo CSV ---
print("üìÇ Selecciona tu archivo CSV con los datos (por ejemplo: vgsales.csv)")
uploaded = files.upload()

# Obtener el nombre del archivo subido
file_name = list(uploaded.keys())[0]

# --- 2Ô∏è‚É£ Cargar el archivo en un DataFrame ---
df = pd.read_csv(file_name)

# Mostrar las primeras filas
print("\n‚úÖ Archivo cargado correctamente:")
print(df.head())

# --- 3Ô∏è‚É£ Verificar existencia de la variable Global_Sales ---
if 'Global_Sales' not in df.columns:
    raise ValueError("‚ùå La columna 'Global_Sales' no se encuentra en el archivo CSV.")

# Convertir a num√©rico en caso de que tenga valores no num√©ricos
df['Global_Sales'] = pd.to_numeric(df['Global_Sales'], errors='coerce')

# --- 4Ô∏è‚É£ Mostrar estad√≠sticas b√°sicas ---
print("\nüìà Estad√≠sticas descriptivas de Global_Sales:")
print(df['Global_Sales'].describe())

# --- 5Ô∏è‚É£ Graficar histograma ---
plt.figure(figsize=(8, 5))
plt.hist(df['Global_Sales'].dropna(), bins=20, edgecolor='black')
plt.title('Histograma de Global_Sales')
plt.xlabel('Ventas Globales (en millones)')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --- 6Ô∏è‚É£ Graficar diagrama de caja ---
plt.figure(figsize=(5, 7))
plt.boxplot(df['Global_Sales'].dropna(), vert=True)
plt.title('Diagrama de Caja de Global_Sales')
plt.ylabel('Ventas Globales (en millones)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
